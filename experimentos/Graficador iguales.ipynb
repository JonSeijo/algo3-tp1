```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
datos_naive_random = pd.read_csv('../datos/naive/iguales.csv')
datos_naive_random.drop(datos_naive_random.columns[50], axis = 1,inplace=True) # Borro columna fea
#datos_naive_random.drop(15, inplace=True) # Borro n=16 para achicar grafico
datos_naive_random['mean'] = datos_naive_random.mean(axis = 1)
datos_naive_random
```

```python
datos_poda_random = pd.read_csv('../datos/poda/iguales.csv')
datos_poda_random.drop(datos_poda_random.columns[50], axis = 1,inplace=True) # Borro columna fea
datos_poda_random['mean'] = datos_poda_random.mean(axis = 1)
datos_poda_random
```

```python
%matplotlib inline
tiempos_naive_random = datos_naive_random['mean']
tiempos_poda_random = datos_poda_random['mean']

# 50 repes
# Naive: n = 15
# Poda: n = 40

plt.clf()
plt.figure(figsize=(8,6))
plot_naive = tiempos_naive_random.plot()
plot_poda = tiempos_poda_random.plot()

plt.show()
```

```python
#ESCALA LOGARITMICA

%matplotlib inline
tiempos_naive_random = datos_naive_random['mean']
tiempos_poda_random = datos_poda_random['mean']

# 50 repes
# Naive: n = 15
# Poda: n = 40

plt.clf()
plt.figure(figsize=(8,6))
plot_naive = tiempos_naive_random.plot(logy = True)
plot_poda = tiempos_poda_random.plot(logy = True)

plt.show()
```
